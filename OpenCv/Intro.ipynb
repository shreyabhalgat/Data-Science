# Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Reading and Displaying Images

```python
img=cv2.imread("./dog.jpg")
```

```python
img.shape
```

```python
type(img)
```

```python
plt.imshow(img) #BGR
```

```python
cv2.imshow("Dog Image",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

## Changing Colorspace

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb) #rgb
```

## Image Resizing

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
img_new=cv2.resize(img_rgb,(500,500))
```

```python
plt.imshow(img_new)
```

## Blur

```python
img_blur=cv2.blur(img_rgb,(20,20))
```

```python
plt.imshow(img_blur)
```

## Image Rotation

```python
img_rgb.shape
```

```python
rows,cols=img_rgb.shape[:-1]
```

```python
rows
```

```python
cols
```

```python
M=cv2.getRotationMatrix2D((rows/2,cols/2),10,0.8)
```

```python
i=cv2.warpAffine(img_rgb,M,(cols,rows))
```

```python
plt.imshow(i)
```

## Image Translation

- Shifting image few pixels UPs or Few pixel down left or right

```python
M=np.float32([[1,0,50],[0,1,20]])
```

```python
M
```

```python
i=cv2.warpAffine(img_rgb,M,(cols,rows))
```

```python
plt.imshow(i)
```

## Edge Detection

- canny edge Detection

```python
i=cv2.Canny(img_rgb,100,100)
plt.imshow(i)
```
