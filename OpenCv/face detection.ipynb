# Face Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# download haarcascade frontalface code

img=plt.imread("./group-photo.jpg")
```

```python
detector=cv2.CascadeClassifier("./haarcascade_frontalface_default.xml")
```

```python
all_faces=detector.detectMultiScale(img,1.3)
```

```python
all_faces.shape
```

```python
x,y,w,h=all_faces[0]
```

```python
img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(img)
```

```python
#plotting reactangle for all images
for face in all_faces:
    x,y,w,h=face
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(img)
```
