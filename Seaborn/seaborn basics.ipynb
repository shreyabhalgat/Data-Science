```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')  
#no need to load dataset from external resource
#seaborn provides internal datasets which can be load using load datatset function
```

```python
iris.head()
```

```python
#plot the distribution of the data 
#umm preety much like histogram
sns.distplot(iris['petal_length'],bins=20)
plt.show()
```

```python
# the line is the graph is known as KDE 
'''
kde stand for kernal density estimation 
you can remove kde by adding paramter kde=false
'''
```

```python
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'],y=iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x= 'petal_length',y='petal_width' ,data=iris,hue='species')
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex')
plt.show()
```

### pairplot

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris,hue='species')
```
