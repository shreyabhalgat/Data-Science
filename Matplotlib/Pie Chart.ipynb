# Pie Chart

- plt.pie()
- plot the pie chart with the provided labels and values

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
subjects=["phy","chem","maths","English"]
marks=np.array([78,45,90,70])
```

```python
plt.pie(x=marks,labels=subjects)
plt.show()
```

```python
plt.pie(x=marks,labels=subjects,explode=[0,1,0,0],shadow=True,autopct="%0.2f")
plt.show()
```
