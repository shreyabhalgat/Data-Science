# Line Plots

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show() #intermiaget o/p of all the charts
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show() #intermiaget o/p of all the charts
```

```python
#customization
plt.plot(x,y1,color='black',marker='o',label="iphone")
plt.plot(x,y2,color='red',marker='*',markersize=10,linestyle='dashed',label="android")

#giving names to the axis
plt.xlabel("time")
plt.ylabel("Price")
plt.title("Prices of smartphones over time")

# to display the names of the lines i.e android and iphone you will have to use the fucntion legend()
plt.legend()
plt.show() #intermiaget o/p of all the charts
```
