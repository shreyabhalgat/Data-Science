# Subplot

- create multiple plots in just single figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10,0.1)
y1=x**2
y2=x**3
y3=np.sin(x)
y4=np.cos(x)
```

```python
x
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```
